```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
folder = "C:/Users/sucod/OneDrive/Lenovo/Documentos/Sql/"

def sublist(lst, n):
    sub=[] ; resultado=[]
    for i in lst:
        sub+=[i]
        if len(sub)==n: resultado+=[sub] ; sub=[]
    if sub: resultado+=[sub]
    return resultado
```

```python
df = pd.read_csv(r"{}".format(folder)+"/CSVLoja/venda_10.csv")
df.tail()
```

```python
df = pd.read_csv(r"{}".format(folder)+"/CSVLoja/venda_10.csv")

sub=sublist(df["vendaValor"].to_numpy(),750)
sub=[sum(x) for x in sub]

plt.plot(sub)
plt.ylabel("Vendas por Período")
```
